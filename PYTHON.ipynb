Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5

```python
def highest_frequency_length(string):
    # Split the string into words
    words = string.split()

    # Create a dictionary to store the frequency of each word
    word_freq = {}
    for word in words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
            
    max_frequency = max(word_freq.values())
    max_length = 0
    for word, frequency in word_freq.items():
        if frequency == max_frequency and len(word) > max_length:
            max_length = len(word)

    return max_length


string1 = "apple apple banana orange banana orange orange"
print(highest_frequency_length(string1)) 

string2 = "cat dog cow"
print(highest_frequency_length(string2))  
```

Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO

```python
def is_valid_string(s):
    char_freq = {} 
    
   
    for char in s:
        if char in char_freq:
            char_freq[char] += 1
        else:
            char_freq[char] = 1

   
    if len(set(char_freq.values())) == 1:
        return "YES"
    
   
    for freq in char_freq.values():
        if list(char_freq.values()).count(freq) == len(char_freq) - 1:
            return "YES"

    return "NO"

print(is_valid_string("abc"))  
print(is_valid_string("abcc"))  

print(is_valid_string("aaabbbccdd"))
```

```python
import requests
import pandas as pd

# API URL
api_url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Send GET request to the API URL and retrieve the response
response = requests.get(api_url)

# Extract relevant data from the response
data = response.json()

# Extract episode data from the embedded episodes
episodes = data["_embedded"]["episodes"]

# Create a DataFrame to store the extracted data
df = pd.DataFrame(episodes)

# Specify the columns to keep
columns_to_keep = ["season", "number", "name", "airdate", "airtime", "runtime", "summary"]

# Filter the DataFrame to keep only the specified columns
df = df[columns_to_keep]

# Convert the airdate column to datetime format
df["airdate"] = pd.to_datetime(df["airdate"])

# Save the DataFrame as a CSV file
df.to_csv("westworld_episodes.csv", index=False)

# Print a message to confirm successful conversion
print("Data has been saved as westworld_episodes.csv")
```

```python
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def count_pos_tags(text):
    # Tokenize the text into words
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

    # Tag the tokens with their respective part-of-speech
    tagged_tokens = nltk.pos_tag(filtered_tokens)

    # Initialize counts
    counts = {'nouns': 0, 'verbs': 0, 'pronouns': 0, 'adjectives': 0}

    # Count the number of each part-of-speech
    for word, tag in tagged_tokens:
        if tag.startswith('N'):
            counts['nouns'] += 1
        elif tag.startswith('V'):
            counts['verbs'] += 1
        elif tag.startswith('P'):
            counts['pronouns'] += 1
        elif tag.startswith('J'):
            counts['adjectives'] += 1

    return counts

# Example usage
text = "The quick brown fox jumps over the lazy dog."
counts = count_pos_tags(text)
print(counts)
```
